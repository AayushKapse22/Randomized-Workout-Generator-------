```python
#generate workout based on selected muscle group and location
def generate_workout():
    selected_muscle_group = muscle_group_var.get()
    selected_location = location_var.get()
    if selected_muscle_group == "" or selected_location == "":
        messagebox.showwarning("Input error","please select both Muscle Group & Location")
        return
        
    exercise_list = exercises[selected_muscle_group][selected_location]
    selected_difficulty = difficulty_var.get()
    if selected_difficulty == "":
        messagebox.showwarning("Input Error", "Please select a difficulty level.")
        return
   # Filter exercises based on selected difficulty
    filtered_exercises = [ex for ex in exercise_list[selected_difficulty] if ex['difficulty'] == selected_difficulty]
    
    random_exercises = random.sample(filtered_exercises, 5)  # Pick 5 random exercises

    workout_text = ""
    for exercise in random_exercises:
        workout_text += f"{exercise['name']} - {exercise['difficulty']} - {exercise['duration']}\n"

    workout_label.config(text=workout_text)
```

```python
# Create the main window
root = tk.Tk()
root.title("Workout Generator")
    
```

```python
# Muscle group selection
muscle_group_label = tk.Label(root, text="Select Muscle Group:")
muscle_group_label.grid(row=0, column=0)
muscle_group_var = tk.StringVar()
muscle_group_menu = tk.OptionMenu(root, muscle_group_var, "chest", "triceps", "biceps", "abs", "legs", "back", "shoulders", "cardio")
muscle_group_menu.grid(row=0, column=1)
```

```python
# Location selection
location_label = tk.Label(root, text="Select Location (Home/Gym):")
location_label.grid(row=1, column=0)
location_var = tk.StringVar()
location_menu = tk.OptionMenu(root, location_var, "home", "gym")
location_menu.grid(row=1, column=1)
```

```python
# Difficulty selection
difficulty_label = tk.Label(root, text="Select Difficulty Level:")
difficulty_label.grid(row=2, column=0)
difficulty_var = tk.StringVar()
difficulty_menu = tk.OptionMenu(root, difficulty_var, "easy", "medium", "hard")
difficulty_menu.grid(row=2, column=1)
```

```python
# Generate workout button
generate_button = tk.Button(root, text="Generate Workout", command=generate_workout)
generate_button.grid(row=3, column=0, columnspan=2)
```

```python
# Display generated workout
workout_label = tk.Label(root, text="", justify="left")
workout_label.grid(row=4, column=0, columnspan=2)
```

```python
# Run the GUI
root.mainloop()
```
